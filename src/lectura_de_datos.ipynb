```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('../data/train.csv')
test_data =pd.read_csv('../data/test.csv')
```

```python
train_info = train_data.info()
train_head = train_data.head()
test_info = test_data.info()
test_head = test_data.head()
print(f'train_info:\n{train_info}\nTrain_head:\n{train_head}\ntest_info:\n{test_info}\nTest_head:\n{test_head}')
```

```python
#Drop colums whith too many missing values or not usefull for prediction
columns_to_drop = ['Id','Alley','PoolQC','Fence','MiscFeature']
train_data_cleaned = train_data.drop(columns=columns_to_drop)
test_data_cleaned = test_data.drop(columns=columns_to_drop)
```

```python
#Impute Missing numerical values with the media 
train_data_cleaned['LotFrontage'].fillna(train_data_cleaned['LotFrontage'].median(), inplace=True)
test_data_cleaned['LotFrontage'].fillna(test_data_cleaned['LotFrontage'].median(), inplace=True)
```

```python
#Impute missing categorical values with the most common Value(mode)
for col in ['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']:
    train_data_cleaned[col].fillna(train_data_cleaned[col].mode()[0], inplace=True)
    test_data_cleaned[col].fillna(test_data_cleaned[col].mode()[0], inplace=True)
```

```python
#impute the remaining numerical columns with the median
train_data_cleaned['MasVnrArea'].fillna(train_data_cleaned['MasVnrArea'].median(), inplace=True)
test_data_cleaned['MasVnrArea'].fillna(test_data_cleaned['MasVnrArea'].median(), inplace=True)
```

```python
#Check if there are any remaining missing values
train_nulls = train_data_cleaned.isnull().sum()
test_nulls = test_data_cleaned.isnull().sum()
print(f'train_nulls:\n{train_nulls}\n test_nulls:\n{test_nulls}')
```

```python
#Identify reamining columns with missing values
train_missing_cols = train_data_cleaned.columns[train_data_cleaned.isnull().any()]
test_missing_cols = test_data_cleaned.columns[test_data_cleaned.isnull().any()]
```

```python
#Display columns with missing values and the count of missing values for both datasets
train_missing_values = train_data_cleaned[train_missing_cols].isnull().sum()
test_missing_values = test_data_cleaned[test_missing_cols].isnull().sum()

print(f'Train_missing_values:\n{train_missing_values}\n Test_missing_values:\n{test_missing_values}')
```

```python
#Impute more missin values for categorical columns 
train_data_cleaned['FireplaceQu'].fillna('NoFireplace', inplace=True)
test_data_cleaned['FireplaceQu'].fillna('NoFireplace', inplace=True)
```

```python
#Garage-related columns: 
garage_columns = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']
for col in garage_columns:
    train_data_cleaned[col].fillna('NoGarage', inplace=True)
    test_data_cleaned[col].fillna('NoGarage', inplace=True)
```

```python
#Inpute GarageYrBlt with 0
train_data_cleaned['GarageYrBlt'].fillna(0, inplace=True)
test_data_cleaned['GarageYrBlt'].fillna(0, inplace=True)
```

```python
#inpute Electrical with the mode
train_data_cleaned['Electrical'].fillna(train_data_cleaned['Electrical'].mode()[0], inplace=True)
```

```python
#inpute missing values in test dataset 
columns_to_impute_mode = ['MSZoning', 'Utilities', 'Exterior1st', 'Exterior2nd', 
                          'KitchenQual', 'Functional', 'SaleType']
for col in columns_to_impute_mode:
    test_data_cleaned[col].fillna(test_data_cleaned[col].mode()[0], inplace=True)
```

```python
#Inpute remaing numerical columns with media
for col in ['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 
            'GarageCars', 'GarageArea']:
    test_data_cleaned[col].fillna(test_data_cleaned[col].median(), inplace=True)
```

```python
train_nulls_after = train_data_cleaned.isnull().sum()
test_nulls_after = test_data_cleaned.isnull().sum()
print(f'train_nulls_after:\n{train_nulls_after}\n test_nulls_after:\n{test_nulls_after}')
```

```python
#Apply Onde-hot encoding
train_data_encoded = pd.get_dummies(train_data_cleaned, drop_first=True)
test_data_encoded = pd.get_dummies(test_data_cleaned, drop_first=True)
```

```python
#Alinear datos
train_data_encoded, test_data_encoded = train_data_encoded.align(test_data_encoded, join='left', axis=1)
```

```python
print(f'train_data_encoded:\n{train_data_encoded}\n test_data_encoded:\n{test_data_encoded}')
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
num_cols = train_data_encoded.select_dtypes(include=['float64', 'int64']).columns
```

```python
scaler = MinMaxScaler()
train_data_encoded[num_cols] = scaler.fit_transform(train_data_encoded[num_cols])
test_data_encoded[num_cols] = scaler.fit_transform(test_data_encoded[num_cols])
```

```python
#Check
print(f'train_data_encoded:\n{train_data_encoded.head()}, test_data_encoded:\n{test_data_encoded.head()}')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = train_data_encoded.drop('SalePrice', axis=1)
y = train_data_encoded['SalePrice']
x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
y_pred = model.predict(x_valid)
```

```python
mse = mean_squared_error(y_valid, y_pred)
print(f"Mean squared Error: {mse}")
```
